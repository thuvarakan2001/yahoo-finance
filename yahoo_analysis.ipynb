```python
#A look into cost of Revenue and Total revenue of Microsoft  
revenue = df.loc['CostOfRevenue':'TotalRevenue']
```

```python
#Format the revenue into readible number format
revenue = revenue / 1000000000
```

```python
revenue 
```

```python
#Transpose the revenue to pivot
revenue = revenue.T
```

```python
#Pivoted Revenue 
revenue
```

```python
#Plot for Microsoft's Revenue analysis looking at total and cost of revenue for microsoft
plt.figure(figsize=(20,10))
x_values = [i.year for i in revenue.index]
data = revenue

#Set X-axis, setting width and multiplier for offsetting each bar in the grouped bar chart 
x = np.arange(len(x_values))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

#Itterating of each revenue category, creates a bar for each attribute with an offset to group them, adds labels on top of the bars, and increments the multiplier to position the next set of bars.
for attribute, measurement in data.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('USD in Billions')
ax.set_xlabel('Year')
ax.set_title(f'{stock} Revenue Analysis')
ax.set_xticks(x + width, x_values)
ax.legend(revenue.columns,loc=9, ncols=3)
ax.set_ylim(0, 500)
plt.show()
```

```python
#Loading top technology stocks 
#Dataset from https://www.sectorspdrs.com/mainfund/xlk
df = pd.read_csv('https://github.com/thuvarakan2001/yahoo-finance/blob/main/index_holdings_xlk.csv?raw=true', header = 1)
```

```python
df.head()
```

```python
stocks = df['Symbol'].to_list()
```

```python
#Creating a list for stocks, a for loop to iterate through the list and calling yahoo finance api ticker to get trailing price to earnings
#Appending all the data from the server into a new list
#Best practice to using try and except when working with APIs
portfolio = []
for co in stocks: 
    try: 
        pe = yf.Ticker(co).get_info()['trailingPE']
        portfolio.append((co,pe))
    except:
        pass
```

```python
#Tech stocks
portfolio
```

```python
#Dataframe containing the entire portfolio with stock signs and Price earning ratio
df = pd.DataFrame(portfolio, columns = ['Stocks', 'PE'])
```

```python
df.head()
```

```python
df['PE'].describe()
```

```python
#Plotting the price-to-earnings ratio of technology stocks from the dataset and look into insights
plt.figure(figsize =(25,10))
plt.scatter(df['Stocks'],df['PE'],s=df['PE'], c=df['PE'],cmap ='viridis')
plt.hlines(df['PE'].mean(),df.index[0],df.index[-1], linestyle =':')
plt.xticks(rotation = 45)
plt.colorbar()
plt.xlabel('Stock Symbols')
plt.ylabel('PE')
plt.title('The price-to-earnings ratio of Tech Stocks')
```

```python
#Getting the valuation of Microsoft
msft_valuation = si.get_stats_valuation('MSFT')
```

```python
msft_valuation
```

```python
#Renaming columns into ratios, using inplace = true to save changes 
msft_valuation.rename(columns={'Unnamed: 0': 'Ratios'}, inplace = True)
```

```python
#A look at microsoft's valuation
msft_valuation
```

```python
#Taking all tickers from dow stocks
dow_list = si.tickers_dow()
```

```python
#Dow stocks in a list
dow_list
```

```python
#Creating new dataframe, column ratios and for loop to iterate through the list of symbols getting statistical valuation for each company and only the first column
stats_df = pd.DataFrame()
stats_df['Ratios'] = msft_valuation['Ratios']
for co in dow_list: 
    stats = si.get_stats_valuation(co)
    column = stats[stats.columns[1]]
    stats_df [co] = column 
```

```python
stats_df.index = stats_df['Ratios']
```

```python
#Transpose the dataframe
stats_df = stats_df.T 
```

```python
#Dropping the column only keeping the column names
stats_df.drop('Ratios',inplace= True)
```

```python
#All company data, but market cap is in trillions and billions so we need to create a new column to convert to everything to billions
stats_df
```

```python
#Applying lambda function to convert the trillions to billions and the datatype to floats 
#To convert a value from trillions to billions, I multiply by 1000 because there are 1000 billions in a trillion
#x will be each value from the column market cap and using strip function to check if it has T for trillions and B for billions and if statement for the logic
stats_df['Market_Cap_Float'] = stats_df['Market Cap'].apply(lambda x: float(x.strip('T').strip()) * 1000 if 'T' in x else (float(x.strip('B').strip()) if 'B' in x else None))
```

```python
#Checking datatype is float
print(stats_df['Market_Cap_Float'].dtype)
```

```python
stats_df
```

```python
#Plotting market cap for each company for further analysis
plt.figure(figsize =(20,8))
plt.scatter(stats_df.index ,stats_df['Market_Cap_Float'], s=stats_df['Market_Cap_Float'], c=stats_df['Market_Cap_Float'],cmap ='viridis')
plt.xticks(rotation = 45)
plt.colorbar()
plt.ylim(0,4000)
plt.xlabel('Stock Symbols')
plt.ylabel('Market Cap in Billions')
plt.title('Market Cap')
```

```python
#Using copy method to slice get Nvidia only
stock_df = mystocks_df[['NVDA']].copy()
stock_df.head()
```

```python
#Calculate Moving Average for 10 day
stock_df['moving_average'] = stock_df.rolling(10).mean()
```

```python
#Dropping NaN values in dataframe
stock_df.dropna(inplace=True)
```

```python
max_price = stock_df[stock_df['NVDA']== stock_df['NVDA'].max()]['NVDA']
max_price
```

```python
min_price = stock_df[stock_df['NVDA']== stock_df['NVDA'].min()]['NVDA']
min_price
```

```python
#Plotting Nvidia's 10 day moving average price and stock price
x = stock_df.index 
y = stock_df['NVDA']
stock = stock_df.columns[0]
plt.figure(figsize =(20,8))
plt.style.use('seaborn-v0_8')
plt.plot(x,y,label =f'{stock} price')
plt.plot(x,stock_df['moving_average'], color = 'red', linestyle = ':', label='10 Day Moving Average')
plt.scatter(max_price.index ,max_price.values, s=600, c='red',marker='*')
plt.scatter(min_price.index ,min_price.values, s=600, c='green',marker='^')
plt.legend(fontsize=12)
plt.title(f'{stock} Daily Prices', fontsize = 22)
plt.ylabel ('USD price', fontsize=22)
plt.xlabel ('Dates', fontsize=22)
plt.xticks(fontsize = 18)
plt.yticks(fontsize = 18)
```
